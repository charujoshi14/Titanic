```python
plt.figure(figsize=(18,12))

ax1=plt.subplot2grid((2,2),(0,0))
train['Age'].plot(kind='kde',color='red',label='train',alpha=0.6)
test['Age'].plot(kind='kde',color='blue',label='test',alpha=0.6)
ax1.set_xlabel("Age")
ax1.set_title("Age distribution")
plt.legend(loc='best')

ax2=plt.subplot2grid((2,2),(0,1))
train['Pclass'].value_counts().plot(kind='bar',color='red',label='train',alpha=0.6)
test['Pclass'].value_counts().plot(kind='bar',color='blue',label='test',alpha=0.6)
ax2.set_xlabel("Pclass")
ax2.set_title("Pclass distribution")
plt.legend(loc="best")

ax3=plt.subplot2grid((2,2),(1,0))
train['Sex'].value_counts().plot(kind='bar',color='red',label='train',alpha=0.6)
test['Sex'].value_counts().plot(kind='bar',color='blue',label='test',alpha=0.6)
ax3.set_xlabel("Sex")
ax3.set_title("Sex distribution")
plt.legend(loc="best")

ax4=plt.subplot2grid((2,2),(1,1))
train['Embarked'].value_counts().plot(kind='bar',color='red',label='train',alpha=0.6)
test['Embarked'].value_counts().plot(kind='bar',color='blue',label='test',alpha=0.6)
ax4.set_xlabel("Embarked")
ax4.set_title("Embarked distribution")
plt.legend(loc="best")
```

```python
plt.figure(figsize=(10,6))


train[train['Survived']==1].Age.plot(kind='kde',label='survived',color='red',alpha=0.6)
train[train['Survived']==0].Age.plot(kind='kde',label='not_survived',color='blue',alpha=0.6)
plt.legend(loc='best')
```
